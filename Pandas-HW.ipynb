```python
import pandas as pd
import numpy as np
```

```python
csvFile = 'purchase_data.json'
data = pd.read_json(csvFile)

data.head()
```

```python
data.columns
```

```python
#Display unique number of players
dict = {"Total Players": [len (data["SN"].unique())] }
pd.DataFrame.from_dict(dict,orient='columns')
```

```python
#Purchasing Analysis
purchaseDict = {"Number of Unique Items": [len (data["Item Name"].unique())],                            
                "Total Revenue": [data["Price"].sum()],
                "Average Price": [data["Price"].mean()],
                "Number Of Purchases": [len (data) ]}
pd.DataFrame.from_dict(purchaseDict, orient = 'columns').reindex(columns=["Number of Unique Items","Average Price","Number Of Purchases","Total Revenue"])
```

```python
#Gender Demographics

genderDict = {}
genderList = []
percentList = []
totalList = []

genderGroup = data.groupby(["Gender"], sort = False)

total = genderGroup["Gender"].count().sum()


for row in genderGroup["Gender"]:
  genderList.append(row[0])
  totalList.append(len(row[1]))
  percentList.append(100 *(len(row[1])/total)) 
                   

genderDict["Gender"] = genderList
genderDict["Percentage of Players"] = percentList
genderDict["Total Count"] =  totalList  

 
myPd = pd.DataFrame.from_dict(genderDict,orient='columns')
myPd.set_index('Gender', inplace = True)

#display output
myPd

```

```python
#Purchasing Analysis (Gender)
purchaseDict = {}

#references
#genderList
#genderGroup

genderPriceGroup = data.groupby(['Gender'], sort = False)

purchaseDict["Gender"] = genderList
purchaseDict["Purchase Count"] = totalList
purchaseDict["Average Purchase"] = genderPriceGroup['Price'].mean()
purchaseDict["Total Purchase Value"] = genderPriceGroup['Price'].sum()
purchaseDict["Normalized totals"] = genderPriceGroup['Price'].mean()

purchasePd = pd.DataFrame.from_dict(purchaseDict, orient ='columns')
purchasePd.set_index('Gender', inplace = True)

#display output
purchasePd
```

```python
#Age Demographics

ageDict = {}
bins= [0, 10, 14,19,24,29,34,39,10000]
nRange = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]
out = []
percent = []
combined = []

ageGroup = data.groupby(['Age'], sort = False)

#Get the Series
occurances = ageGroup["Age"].count()

keyV = occurances.keys()
occ = occurances.get_values()

#get the worked from bin for histgram
for x in range(0, len(keyV)):
    for dumb in range(0, occ[x]):
         combined.append(keyV[x])
        
out,leftover =  np.histogram(combined, bins=bins)

#do percentages
for x in range(0,len(out)):
    percent.append( (out[x]/out.sum()) * 100)
    

ageDict["nRange"] = nRange 
ageDict["Total Count"] = out
ageDict["Percentage Of Players"] =  percent


agePd = pd.DataFrame.from_dict(ageDict,orient ='columns')
agePd.set_index('nRange', inplace = True)
agePd
```
